# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
Base.prepare(engine, reflect=True)

# reflect the tables

# from sqlalchemy import inspect
# inspector = inspect(engine)
# inspector.get_table_names()

#---returns ['measurement', 'station']
#----------------------------------------------------------------------------------
# columns = inspector.get_columns('measurement')
# for c in columns:
#     print(c['name'], c["type"])
    
#---returns 
# id INTEGER
# station TEXT
# date TEXT
# prcp FLOAT
# tobs FLOAT
#----------------------------------------------------------------------------------
# columns = inspector.get_columns('station')
# for c in columns:
#     print(c['name'], c["type"])
    
#---returns 
# id INTEGER
# station TEXT
# name TEXT
# latitude FLOAT
# longitude FLOAT
# elevation FLOAT

```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
max_date = session.query(Measurement.date, Measurement.station).order_by(Measurement.date.desc()).limit(1)
print(f"The max date is {max_date[0][0]}")

# Perform a query to retrieve the data and precipitation scores
date_compare = dt.datetime(2016, 8, 23)
dates_12mo = session.query(Measurement.date, func.avg(Measurement.prcp)).\
    filter(Measurement.date>=date_compare).\
    group_by(Measurement.date).\
    order_by(Measurement.date).\
    all()


# Save the query results as a Pandas DataFrame and set the index to the date column
dates_12mo_df = pd.DataFrame(dates_12mo, columns=['date','prcp'])
dates_12mo_df.set_index('date', inplace=True)


# Sort the dataframe by date
dates_12mo_df.sort_values(by=['date'])


# Use Pandas Plotting with Matplotlib to plot the data
bar_plot = dates_12mo_df.plot.bar(figsize=(13,5), title = "Average Precipitation by Day", rot=45, legend=False)
xticks = bar_plot.xaxis.get_major_ticks()
for i,tick in enumerate(xticks):
    if i%8 != 0:
        tick.label1.set_visible(False)
bar_plot.set_xlabel("Date of Observation")
bar_plot.set_ylabel("Precipitation Level")
bar_plot.xaxis.grid(False, which='minor')
bar_plot.xaxis.grid(False, which='major')
plt.tight_layout()
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
avg_prcp = dates_12mo_df["prcp"].mean()

print(f"The average precipitation is {round(avg_prcp,2)}")
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Measurement.station).group_by(Measurement.station).count()


print(f"There are {station_count} stations")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
station_active = session.query(Measurement.station, func.count(Measurement.date)).group_by(Measurement.station).order_by(func.count(Measurement.date).desc())

for each in station_active:
    print(f"The station {each[0]} has {each[1]} active records.")
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
temps = session.query(func.min(Measurement.tobs),func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
    filter(Measurement.station == "USC00519281").all()

print(f"The lowest recorded temp was {temps[0][0]} degrees")
print(f"The highest recorded temp was {temps[0][1]} degrees")
print(f"The average recorded temp was {round(temps[0][2],1)} degrees")


```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
station_active = session.query(Measurement.station,func.max(Measurement.date), func.count(Measurement.tobs)).group_by(Measurement.station).order_by(func.count(Measurement.date).desc()).limit(1)
print(f"The station with the most TOBS measurements is {station_active[0][0]}, the last date of measurement is {station_active[0][1]}")

last_date = dt.datetime(2016, 8, 18)
most_active_station = session.query(Measurement.date, func.avg(Measurement.tobs)).\
    filter(Measurement.date>=last_date).\
    group_by(Measurement.date).\
    order_by(Measurement.date).\
    all()

mas_df = pd.DataFrame(most_active_station, columns=['date','tobs'])
mas_df.set_index('date', inplace=True)
mas_df.sort_values(by=['date'])

hist_plot = mas_df.plot.hist(figsize=(13,5), legend = False, title = "Histogram of Temps")
hist_plot.set_xlabel("Temperature")
plt.tight_layout()
plt.show()





```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
